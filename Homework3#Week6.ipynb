<h2>Soal 6: Membuat Function dengan Multiple arguments</h2>

Buatlah suatu fungsi untuk melakukan penggabungan antara dua list

```python
obj_list = [2, 4, 5, 6]
obj_penambah = [1, 2, 3]

#yourcode
def gabung_list(obj_list, obj_penambah):
    # isikan kode
    obj_list.extend(obj_penambah)
    return obj_list
    
print(gabung_list(obj_list, obj_penambah))
```

Expected Output:

[2, 4, 5, 6, 1, 2, 3]

<h2>Soal 7: Exercise 1</h2>

buat fungsi untuk menghitung selisih kuadrat (x kuadrat - y kuadrat) dari dua bilangan

```python
def get_selisih_kuadrat(x,y):
    x = (x ** 2)
    y = (y ** 2)
    z = x - y
    return z

print(get_selisih_kuadrat(7,2))
print(get_selisih_kuadrat(10,8))
print(get_selisih_kuadrat(-5,6))
```
