### Fufun Dany Artheja

```python
import numpy as np
x = np.arange(start=1, stop=31)
x = x.reshape((6, 5))
print(x)


```

Homework - Introduction to Numpy.png

```python
#Blue Box Syntax
blue = x[2:4 , 0:2]
print(blue)
```

```python
#Green Box Syntax
y = np.diag(x, k=1)
Green = y.reshape(-1,1)
print(Green)
```

```python
#Red Box Syntax
Red1 = x[:1 , 3:]
Red2 = x[4: , 3:]
Red = np.concatenate((Red1, Red2))
print(Red)
```
