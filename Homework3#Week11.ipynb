```python
#Import Libraries
import pandas as pd
import numpy as np
```

```python
#Import Dataset
SourceFile = '/content/drive/MyDrive/Homework - Data Preprocessing.csv'
df = pd.read_csv(SourceFile)
print(df)
print(df.info())
```

```python
#Menghitung row yang duplicate dan jika ada di drop
print(df.duplicated().sum())
```

```python
#menghitung jumlah nilai nan/null dalam dataset
print(df.isna().sum())
print(df.shape)
print(df.describe())
```

```python
#Mengisi data yang kosong untuk kolom age (numerikal) dengan mean
df['Age'].fillna((df['Age'].mean()), inplace=True)
```

```python
#Mengisi data yang kosong untuk kolom worth (numerikal) dengan median
df['Worth'].fillna((df['Worth'].median()), inplace=True)
```

```python
#Mengisi data yang kosong untuk kolom animal (categorical) dengan mode
df["Animal"].fillna(df["Animal"].mode()[0], inplace=True)
```

```python
# Memecah dataset menjadi data independent dan dependen variabel
X = df[['Animal', 'Age', 'Worth']].values
y = df['Friendly'].values
print(X)
print(y)
```

```python
# Merubah nilai target variable menjadi 0 dan 1
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```python
# Memecah dataset menjadi set training dan test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```
