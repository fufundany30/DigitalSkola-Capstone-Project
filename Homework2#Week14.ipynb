## Getting the Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
```

```python
SourceFile = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/haberman.csv'
df = pd.read_csv(SourceFile, names=['Age','Year of Operation','Number of PAN','Survival Status'])
df.head()
```

## No.1

Bagi kedua data ini menjadi data training dan data test dengan test_size=0.25.

```python
# Memisahkan dataset menjadi predictor dan target
X = df.drop("Survival Status", axis=1)
y = df['Survival Status']

print("X:", X.shape)
print("y:", y.shape)
```

```python
#memisahkan dataset train dengan test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state=27)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler= StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## No.2

Pelajari tentang metrics roc_auc_score kemudian buatlah model dan evaluasi dengan menggunakan teknik cross-validation dengan scoring 'roc_auc'.

```python
# model
decision_tree = DecisionTreeClassifier() 
model_dt = decision_tree.fit(X_train, y_train)

#evaluasi
r_a_score = cross_val_score(decision_tree, X, y, scoring='roc_auc', cv=10)
```

## No.3

Berapa score rata-rata dari model dengan teknik cross-validation tersebut?

```python
#score
print("Score Mean:", round(r_a_score.mean(),2))
```

## No.4

Prediksi data test dengan model yang telah kalian buat.

```python
from sklearn.metrics import accuracy_score
y_test_pred = model_dt.predict(X_test)
test = accuracy_score(y_test, y_test_pred)
print("acc_score:", round(test, 2))
```

## No.5

Bagaimana hasil confusion matrix dari hasil prediksi tersebut?

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_test_pred)
```

## No.6

Bagaimana classification report dari hasil prediksi tersebut?

```python
from sklearn.metrics import precision_score, recall_score, f1_score

print("Precision:", round(precision_score(y_test, y_test_pred),2))
print("Recall:", round(recall_score(y_test, y_test_pred),2))
print("F1_score:", round(f1_score(y_test, y_test_pred),2))
```
