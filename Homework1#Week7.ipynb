# **Write and Read File in Python**

Write :
- r	= Opens a file for reading. (default)
- w	= Opens a file for writing. Creates a new file if it does not exist or truncates the file if it exists.
- x	= Opens a file for exclusive creation. If the file already exists, the operation fails.
- a	= Opens a file for appending at the end of the file without truncating it. Creates a new file if it does not exist.
- t	= Opens in text mode. (default)
- b	= Opens in binary mode.

Read :
- Read() = Reads all char from a file.
- Readline() = Reads all char up to and including \n.
- Readlines() = Reads all line



```python
#Example

f = open("test.txt", "r")
print(f.read())
```

# **Create Table**

```python
conn = psycopg2.connect(
        host = "localhost",
        port = portnumber,
        database= "yourdatabase",
        user = "yourusername",
        password = "yourpassword")

#creating object using the cursor method
cursor = conn.cursor()

cursor.execute("DROP TABLE IF EXISTS sandbox.batch_2.bus")

sql = '''
      create table sandbox.batch_2.bus (
      id int,
      origin varchar,
      destination varchar,
      time time
)'''

cursor.execute(sql)
print("Table created successfully..")

conn.commit()

conn.close()
```

# **Insert Data**

```python
conn = psycopg2.connect(
        host = "localhost",
        port = portnumber,
        database= "yourdatabase",
        user = "yourusername",
        password = "yourpassword")

#creating object using the cursor method
cursor = conn.cursor()

sql ='''insert into sandbox.batch_2.bus values 
        (100, 'Jakarta', 'Bandung', '10:00'),
        (200, 'Jakarta', 'Palembang', '11:00'),
        (300, 'Jakarta', 'Jogja', '12:00')'''

cursor.execute(sql)

conn.commit()
print("Data inserted successfully..")

conn.close()
```

# **Select Data**

```python
conn = psycopg2.connect(
        host = "localhost",
        port = portnumber,
        database= "yourdatabase",
        user = "yourusername",
        password = "yourpassword")

#creating object using the cursor method
cursor = conn.cursor()

sql ='''select * from sandbox.batch_2.bus'''

cursor.execute(sql)

result = cursor.fetchall();
print(result)

conn.commit()

conn.close()
```

# **Update Table**

```python
conn = psycopg2.connect(
        host = "localhost",
        port = portnumber,
        database= "yourdatabase",
        user = "yourusername",
        password = "yourpassword")

#creating object using the cursor method
cursor = conn.cursor()

print('Data sebelum diupdate')
sql = '''
      select * from sandbox.batch_2.bus
'''
cursor.execute(sql)
print(cursor.fetchall())


print('Data sesudah diupdate')
sql = '''
      update sandbox.batch_2.bus
      set id = 100, origin = 'Medan', destination = 'Bandung', time = '10:00'
      where id = 100 
    '''
cursor.execute(sql)

sql = '''
      select * from sandbox.batch_2.bus
'''
cursor.execute(sql)
print(cursor.fetchall())

conn.commit()

conn.close()
```

# **Delete Data**

```python
conn = psycopg2.connect(
        host = "localhost",
        port = portnumber,
        database= "yourdatabase",
        user = "yourusername",
        password = "yourpassword")

#creating object using the cursor method
cursor = conn.cursor()

print('Data sebelum didelete')
sql = '''
      select * from sandbox.batch_2.bus
'''
cursor.execute(sql)
print(cursor.fetchall())

print('Data sesudah didelete')
sql = '''
      delete from sandbox.batch_2.bus where id = 100
    '''
cursor.execute(sql)

sql = '''
      select * from sandbox.batch_2.bus
'''
cursor.execute(sql)
print(cursor.fetchall())

conn.commit()

conn.close()
```
