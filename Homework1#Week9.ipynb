```python
import pandas as pd
import numpy as np

data = {
    'Tipe Ujian' : ['UTS1','UAS1'] * 3,
    'Mata Pelajaran' : ['Statistika','Fisika','Fisika','Statistika','Fisika','Statistika'],
    'Jumlah Peserta' : [14,17,12,16,14,13]
}
df = pd.DataFrame(data).set_index('Tipe Ujian')
df
```

```python
df.pivot_table(index='Tipe Ujian',columns='Mata Pelajaran', aggfunc=np.sum, fill_value=0)
```

```python
import pandas as pd

Data = {
    'Hari' : ['Sabtu','Minggu'],
    'Kebun Binatang' : [271,399],
    'Dufan' : [501,700],
    'Disney' : [1000,1001],
    'Bali' : [900,803]
}
df1 = pd.DataFrame(Data).set_index('Hari')
df1
```

```python
cols = df1.columns
df1.melt(value_vars=cols, var_name='Tempat Hiburan', value_name='Pengunjung', ignore_index=False)

```
