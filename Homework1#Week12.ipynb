## Data Cleaning

### Lower Case

```python
#membuat semua text yang ada di kolom texts menjadi kecil
df['texts'] = df['texts'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

### Removing Punctuation

```python
df['texts'] = df['texts'].str.replace('[^\w\s]','') #menghilangkan tanda baca
df
```

### Removing Number

```python
df['texts'] = df['texts'].str.replace(r'\d+','') #menghilangkan numerik
df
```

### Removing Whitepace

```python
df['texts'].str.strip() #menghilangkan whitespace
df
```

### Removing Hyperlink

```python
df['texts'] = df['texts'].str.replace('http\S+|www.\S+','') # menghilangkan hyperlink
df
```

### Removing Symbol

```python
df['texts'] = df['texts'].str.replace('#[A-Za-z0-9]+','') # menghilangkan symbol hashtag (#)
df['texts'] = df['texts'].str.replace('@[A-Za-z0-9]+','') # menghilangkan symbol mention (@)
df
```

## Tokenization

### Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
st = factory.create_stemmer()

df['texts'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

### Stopword Removal

```python
from nltk.corpus import stopwords

stop = set(stopwords.words('indonesian'))

df['texts'] = df['texts'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df
```

### Split Method

```python
pisah = df['texts'].str.split() #memisahkan kalimat menjadi per kata
print(pisah)
```

### Word Tokenize Method

```python
# import word_tokenize dari modul nltk
from nltk.tokenize import word_tokenize 
 
tokens = df['texts'].apply(word_tokenize)
print(tokens)
```

## Vectorization

### Bag of Words Method

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
text_counts= cv.fit_transform(df["texts"])
pd.DataFrame(text_counts.toarray(), columns = cv.get_feature_names())
```

### TF-IDF Method

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer()
text_counts= cv.fit_transform(df["texts"])
pd.DataFrame(text_counts.toarray(), columns = cv.get_feature_names())
```
